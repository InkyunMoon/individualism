# 라이브러리 임포트

```python
import pandas as pd
import numpy as np
import geopandas as gpd
import json
import seaborn as sns
import matplotlib.pyplot as plt
```

## 서울시 shp 파일 읽어오기

```python
# http://www.gisdeveloper.co.kr/?p=2332 에서 받은 읍면동 단위 shp 파일 사용

shaperaw = r'파일위치\EMD_202005\EMD.shp'
geodf = gpd.read_file(shaperaw)
```

```python
# EMD_CD 필드에서 시작이 11인 데이터가 서울데이터

geodf = geodf[geodf['EMD_CD'].astype(int) <= 11999999]
```

```python
geodf
```

```python
geodf.plot(figsize=(16,8))
```

## 좌표계 설정하기

```python
# 좌표계 통일하기 EPSG:5186

geodf = geodf.to_crs('EPSG:5186')
```

```python
# txt2shp 를 사용하여 shp 파일 생성

moonhwa = r'C:\Users\Soraidaaaaa\pyt\Seoul\Data\문화체육시설\문화체육시설.shp'
```
