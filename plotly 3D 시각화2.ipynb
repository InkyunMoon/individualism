```python
def k_means(data, clusters, random_state):
    km = KMeans(n_clusters=clusters, init='k-means++', n_init=100, max_iter=300, tol=0.0001, random_state = random_state)
    km.fit(data)
    return np.hstack((data, km.predict(data).reshape(-1,1)))

def k_medoids(data, clusters, random_state):
    km = KMedoids(n_clusters = clusters,
                    init='k-medoids++',
                    max_iter = 300,
                    random_state=random_state)
    km.fit(data)
    return np.hstack((data, km.predict(data).reshape(-1,1)))
```

```python
def dbscan(data, eps,min_samples,metric='euclidean'):
    db = DBSCAN(eps=eps, min_samples=min_samples)
    pred = db.fit_predict(np.array(data))
    data['cluster'] = db.fit_predict(data)
    return data
```
