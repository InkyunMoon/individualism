```python
from plotly.offline import plot
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)

import numpy as np
import pandas as pd
import pickle
# !pip install plotly
import plotly
from sklearn.cluster import KMeans, DBSCAN
```

```python
with open('C:/Users/moon/Documents/final_p/after_submit/BASE.bin', 'rb') as f:
    temp = pickle.load(f)
    
temp.columns
df = temp[['GU_CD', 'GU_NM', 'DONG_CD', 'DONG_NM', 'geometry', 'INCOME_RATE', 'AREA_RATE', 'BODY']]

data = df[['INCOME_RATE','AREA_RATE','BODY']]
```

```python
def dbscan(data, eps,min_samples,metric='euclidean'):
    db = DBSCAN(eps=eps, min_samples=min_samples)
    pred = db.fit_predict(np.array(data))
    data['cluster'] = db.fit_predict(data)
    return data
```

```python
temp_df = dbscan(data, 0.2, 5)
```

```python
def graph_3d(df): #1,2,3열에 데이터, 4열에 클러스터인 배열을 인풋으로 받는다.
    data = df.values
    data = go.Scatter3d(
        x=data[:,0],
        y=data[:,1],
        z=data[:,2],
        text = ['point #{}'.format(i) for i in range(data.shape[0])],
        mode='markers',
        marker=dict(
            size=3,
            color=data[:,3],
            colorscale='Jet',
            line=dict(
                #color='rgba(217, 217, 217, 0.14)',
                #color='rgb(217, 217, 217)',
                width=0.0
            ),
            opacity=0.8
        )
    )
    layout = go.Layout(
        autosize=False,
        width=750,
        height=750,
        margin=go.Margin(
            l=50,
            r=50,
            b=100,
            t=100,
            pad=4
        ),
        #paper_bgcolor='#7f7f7f',
        #plot_bgcolor='#c7c7c7'
    )
    fig = go.Figure(data=[data], layout=layout)
    iplot(fig)
```

```python
graph_3d(dbscan(temp_df, 0.6, 5))
```

```python
df = dbscan(temp_df, 0.6, 5)
```

```python
t = go.Scatter3d(x=df['INCOME_RATE'],
                y=df['AREA_RATE'],
                z=df['BODY'],
                mode='markers',
                marker=dict(size=3,color=df['cluster']))
data = [t]
layout = go.Layout(margin = dict(l=0, r=0, b=0, t=0))
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
